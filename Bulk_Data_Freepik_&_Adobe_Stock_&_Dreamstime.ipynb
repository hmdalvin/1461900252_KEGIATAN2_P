```python
import copy
import spacy
import textwrap
import nltk
from rake_nltk import Rake
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def cut_paragraph(paragraph, max_length=100):
    sentences = nltk.sent_tokenize(paragraph)

    result = ""
    current_length = 0

    for sentence in sentences:
        if current_length + len(sentence) > max_length:
            break
        result += sentence + " "
        current_length += len(sentence) + 1

    return result.strip()

def cut_paragraph_manual(paragraph, max_length=100):
    if len(paragraph) > max_length:
        return paragraph[:max_length]
    return paragraph

nlp = spacy.load("en_core_web_sm")

file_path = 'freepik-csv-metadata.xlsx'
df = pd.read_excel(file_path)

dict_data = df.to_dict('dict')

num_columns = len(dict_data['Prompt'])
print("Jumlah kolom dalam dictionary:", num_columns)

for i in range(num_columns):
    text = dict_data['Prompt'][i]
    doc = nlp(text)
    keywords = [chunk.text for chunk in doc.noun_chunks]

    dict_data['Keywords'][i] = ', '.join(keywords)

    rake = Rake()
    rake.extract_keywords_from_text(text)
    keywordsRake = rake.get_ranked_phrases()

    dict_data['Keywords'][i] = ', '.join(keywordsRake)

    paragraph = text

    shortened_paragraph = cut_paragraph(paragraph, 100)
    if not shortened_paragraph:
        shortened_paragraph = cut_paragraph_manual(paragraph)

    dict_data['Title'][i] = shortened_paragraph
```

```python
df = pd.DataFrame(dict_data)

csv_file = 'output_file.csv'
df.to_csv(csv_file, index=False)

# files.download(csv_file)
```

```python
dict_data_adobe_stock = copy.deepcopy(dict_data)
del dict_data_adobe_stock['Prompt']
del dict_data_adobe_stock['Model']
dict_data_adobe_stock = {('Filename' if key == 'File name' else 'Title' if key == 'Title' else 'Keywords' if key == 'Keywords' else key): value for key, value in dict_data_adobe_stock.items()}
dict_data_adobe_stock['Category'] = {}
dict_data_adobe_stock['Release'] = {}
print(dict_data_adobe_stock)
```

```python
df_adobe_stock = pd.DataFrame(dict_data_adobe_stock)

csv_file = 'output_file_adobe_stock.csv'
df_adobe_stock.to_csv(csv_file, index=False)

# files.download(csv_file)
```

```python
dict_data_dreamstime = copy.deepcopy(dict_data)
del dict_data_dreamstime['Model']
dict_data_dreamstime['Category 1'] = {}
dict_data_dreamstime['Category 2'] = {}
dict_data_dreamstime['Category 3'] = {}
dict_data_dreamstime['Free'] = {}
dict_data_dreamstime['W-EL'] = {}
dict_data_dreamstime['P-EL'] = {}
dict_data_dreamstime['SR-EL'] = {}
dict_data_dreamstime['SR-Price'] = {}
dict_data_dreamstime['Editorial'] = {}
dict_data_dreamstime['MR doc Ids'] = {}
dict_data_dreamstime['Pr Docs'] = {}
dict_data_dreamstime = {('Filename' if key == 'File name'
                          else 'Image Name' if key == 'Title'
                          else 'keywords' if key == 'Keywords'
                          else 'Description' if key == 'Prompt'
                          else key): value for key, value in dict_data_dreamstime.items()}

filename_key_count = len(dict_data_dreamstime['Filename'])
for i in range(0, filename_key_count):
    dict_data_dreamstime['Category 1'][i] = '0'
    dict_data_dreamstime['Category 2'][i] = '0'
    dict_data_dreamstime['Category 3'][i] = '0'
    dict_data_dreamstime['Free'][i] = '0'
    dict_data_dreamstime['W-EL'][i] = '1'
    dict_data_dreamstime['P-EL'][i] = '1'
    dict_data_dreamstime['SR-EL'][i] = '0'
    dict_data_dreamstime['SR-Price'][i] = ''
    dict_data_dreamstime['Editorial'][i] = ''
    dict_data_dreamstime['MR doc Ids'][i] = ''
    dict_data_dreamstime['Pr Docs'][i] = ''


ordered_keys = ['Filename', 'Image Name', 'Description',
                'Category 1', 'Category 2', 'Category 3',
                'keywords', 'Free', 'W-EL', 'P-EL', 'SR-EL',
                'SR-Price', 'Editorial', 'MR doc ds', 'Pr Docs']
ordered_dict_data_dreamstime = {key: dict_data_dreamstime[key] for key in ordered_keys if key in dict_data_dreamstime}
print(ordered_dict_data_dreamstime)
```

```python
df_dreamstime = pd.DataFrame(ordered_dict_data_dreamstime)

csv_file = 'output_file_dreamstime.csv'
df_dreamstime.to_csv(csv_file, index=False)

# files.download(csv_file)
```
